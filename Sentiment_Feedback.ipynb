```python
import pandas as pd
import numpy as np
df = pd.read_csv('/content/data_11.csv')
```

```python
df['sentences'] = df['sentences'].fillna("None")
```

```python
from transformers import pipeline

classifier = pipeline(
    "sentiment-analysis",
    model="nlptown/bert-base-multilingual-uncased-sentiment",
    return_all_scores=True,device=0
)
```

```python
batch_size = 64  # safe for Colab GPU
texts = df["sentences"].tolist()

results = []
for i in range(0, len(texts), batch_size):
    batch = texts[i:i+batch_size]
    results.extend(classifier(batch))
```

```python
def extract_sentiment(scores):
    # 'scores' is a list of dictionaries, e.g., [{'label': 'LABEL_0', 'score': 0.006}, ...]
    # Find the dictionary with the highest score
    best = max(scores, key=lambda x: x["score"])
    return best["label"], round(best["score"], 3)

df[["sentiment", "sentiment_score"]] = [
    extract_sentiment(r) for r in results
]
```

```python
df.head(10)
```

```python
def categorize_sentiment(score,sentiment):
    if (score>=0.50 and sentiment=="4 stars") or sentiment == "5 stars":
        return "Positive"
    elif sentiment == "1 star" or (score<=0.50 and sentiment=="3 stars") or sentiment == "2 stars":
        return "Negative"
    else:
        return "Neutral"

df['sentiment_label'] = df.apply(lambda row: categorize_sentiment(row['sentiment_score'], row['sentiment']), axis=1)
# Create or overwrite a column 'sentiment_label'
df.loc[df['topic_group'] == 'Nothing, No Suggestion', 'sentiment_label'] = 'Neutral'
```

```python
df.loc[df['topic_group'] == 'Overall Positive Feedback', 'sentiment_label'] = 'Positive'
```

```python
df.to_csv('df_senti.csv', index=False)
```
