#Merging sector data to master data

*   In sheet_name 'Courses' changed ID to COURSE_ID and SUBOU to SUBOU_ID
*   In sheet_name 'OU' ID changed to SUBOU_ID

All of these changes are made in excel

```python
metadata_course = pd.read_excel('/content/DATASET FOR DSA (1).xlsx', sheet_name='Courses')
```

```python
metadata_course
```

```python
metadata_course = metadata_course[['COURSE_ID', 'SUBOU_ID']] # if other columns are needed do add it here
```

```python
mergeddata1 = data1.merge(
    metadata_course,
    on="COURSE_ID",
    how="left"
)
```

```python
mergeddata1.shape
```

```python
metadata_ou = pd.read_excel('/content/DATASET FOR DSA (1).xlsx', sheet_name='OU')
```

```python
metadata_ou
```

```python
mergeddata2 = mergeddata1.merge(
    metadata_ou,
    on="SUBOU_ID",
    how="left"
)
```
